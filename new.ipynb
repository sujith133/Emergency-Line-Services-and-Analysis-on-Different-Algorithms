```python
def limiting(points_list):    
    for i in range(len(points_list)):
        for j in range(2,len(points_list[i])):
            if(points_list[i][j]>4845):
                points_list[i][j]=points_list[i][j]-4845
    print("limits filter check done")
    return(points_list)
#points_list=limiting(points_list)
```

```python
#print("before chipping: ",len(d))
def clipping(d):
    zeross=0
    for i in range(len(d)):
        for j in range(i,len(d)):
            if(i!=j and d[i]==d[j] and d[i]!=0 and d[j]!=0):
                d[j]=0
                zeross +=1
    #print(zeross)
    for i in range(0,zeross):
        d.remove(0)
    #print("after clipping: ",len(d))
    print("clipping done")
    return (d)
#d=clipping(d)
```

```python
def networking(linkedlist,frm,to):
    G = nx.Graph()
    G.add_weighted_edges_from(linkedlist)
    
    #print("Dijkstra")
    
    dij=nx.dijkstra_path(G, frm, to)
    #print("A-star")
    astr=nx.astar_path(G,frm, to)
    #print("bellman-ford")
    bllman=nx.single_source_bellman_ford(G,frm, to)
    print("done")
    return(dij,astr,bllman)
#dij,astr,bllman=networking(linkedlist,frm, to)
```
